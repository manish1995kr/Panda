```python
df = pd.DataFrame({"key":[1,2,3,4],"A":[1,2,3,4],"B":[1,2,3,4]},index = [0,1,2,3])
df
```

```python
df1 = pd.DataFrame({"key":[1,2,3,4],"A":[5,6,7,8],"B":[5,6,7,8]},index = [0,1,2,3])
df1
```

```python
df_merge = pd.merge(df,df1,how = "outer",on = ["A","B"])
```

```python
df_merge
```

```python
df_merge = pd.merge(df,df1,how = "outer",on = ["A","B"],suffixes = ["_df","_df1"])
```
